**Our aim is predict the employee attrition. It is important to see which variables are contibuting the most in  attrition.**

We check which variables are correlated using a heatmap.
For this we form a dataset of only numerical variables. 

```

hr_data_num = data[['Age','DailyRate','DistanceFromHome', 'Education','EmployeeNumber', 
                       'EnvironmentSatisfaction', 'HourlyRate',                     
                       'JobInvolvement', 'JobLevel','JobSatisfaction','MonthlyIncome', 
                        'MonthlyRate', 'NumCompaniesWorked','PercentSalaryHike', 'PerformanceRating', 
                       'RelationshipSatisfaction', 'StockOptionLevel',
                       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
                       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
                       'YearsWithCurrManager']]
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize= (20,20))
sns.heatmap(hr_data_num.corr(),annot=True)
```

**Make a dataframe of categorical data**

```

hr_data_cat = data[['Attrition', 'BusinessTravel','Department',
                       'EducationField','Gender','JobRole','MaritalStatus',
                       'Over18', 'OverTime']]
hr_data_cat.head()
```

**We convert the Yes/No values of attrition column to 1/0**

```
hr_data_cat["Attrition"]=hr_data_cat["Attrition"].replace("Yes","1")

```

```
hr_data_cat["Attrition"]=hr_data_cat["Attrition"].replace("No","0")
```

```
hr_data_cat.head()
```

**Make dummy columns using pandas get_dummy for the categorical variables**

```
hr_data_cat = pd.get_dummies(hr_data_cat)
hr_data_cat.head()
```

**Delete one of the dummy columns of Attrition**

For Attrition also get dummy creates 2 columns but we need only one column fir it. So we delete one cloumn.

```
hr_data_cat.drop("Attrition_0",axis=1,inplace=True)
```

```
hr_data_cat.head()
```

**Concat the categorical and numerical dataframe into one final dataframe.**

```
hr_data_final = pd.concat([hr_data_num, hr_data_cat], axis=1)
hr_data_final.head()
```

**Split the dataset into training and testing part. In X we take the all the attributes except "Attrition" in y we take only the "Attrition".**

```
from sklearn.model_selection import train_test_split

y = hr_data_final['Attrition_1']
X = hr_data_final.drop('Attrition_1', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

**Applying Random Forest and training the model**

```
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train,y_train)
```

**Predicting the result and checking of accuracy.**

```
from sklearn.metrics import accuracy_score
pred = model.predict(X_test)
accuracy_score(y_test, pred)
```

**A graph representing the top reasons for employee to leave the organization**

```
#Main reasons for leaving the job
features=pd.Series(model.feature_importances_)
#There are many features we are selecting the top 20 reasons
features = features.nlargest(20)
features.plot.bar()



```

**Using Decision Tree for training the model**

```
from sklearn.tree import DecisionTreeClassifier  
classifier = DecisionTreeClassifier()  
classifier.fit(X_train, y_train) 
```

**Predicting the output and checking the accuracy score**

```
y_pred = classifier.predict(X_test) 
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

**Displaying the confusion matrix and classification report**

```
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred))
```
